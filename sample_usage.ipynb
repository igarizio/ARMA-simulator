```python
import matplotlib.pyplot as plt
from ARMA import ARMA
```

```python
phis = [0.1, 0.2, 0.3]
thetas = [0.1, 0.2, 0.5, 0.05]
n = 300
```

```python
plt.title("ARMA")
plt.plot(ARMA.generate_arma(n, phis, thetas, mu=0))
plt.show()

plt.title("AR")
plt.plot(ARMA.generate_ar(n, phis))
plt.show()

plt.title("MA")
plt.plot(ARMA.generate_ma(n, thetas, mu=0))
plt.show()

plt.title("White noise")
plt.plot(ARMA.generate_wn(n))
plt.show()
```
